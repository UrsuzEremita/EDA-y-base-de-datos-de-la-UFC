Los valores NaN en ranking son valores en donde los peleadores todavia no tienen un rango el la ufc.

```python
metacritic['ufc_draws']
```

Los valores NaN en ufc_draws son peleadores que no han peleas empatadas dentro de la ufc.

```python
metacritic['ufc_no_contests']
```

Los valores NaN en ufc_no_contests son peleadores que han tenido peleas sin resultado en la ufc.

Las otra variables (mma_wins, mma_loses, mma_draws y mma_no_contests) son peleadores con peleas ganadas, perdidas, empates y sin resultado (respectivamente) en cualquier otra organizacion profesional de MMA.

```python
# En rankings se les cambiara el valor de NaN a No rank (Sin rango)

metacritic["ranking"].fillna("No rank", inplace = True)
```

```python
# En ufc_draws se les cambiara el valor de NaN a 0

metacritic["ufc_draws"].fillna(0, inplace = True)
```

```python
# En ufc_no_contests se les cambiara el valor de NaN a 0

metacritic["ufc_no_contests"].fillna(0, inplace = True)
```

```python
# Y para las variables mma_wins, mma_loses, mma_draws y mma_no_contests se les cambiara el valor de NaN a 0

metacritic["mma_wins"].fillna(0, inplace = True)
metacritic["mma_loses"].fillna(0, inplace = True)
metacritic["mma_draws"].fillna(0, inplace = True)
metacritic["mma_no_contests"].fillna(0, inplace = True)
```

# Analisis Exploratorio de Datos

```python
alt.Chart(metacritic).mark_circle(size=60).encode(
    alt.X('ufc_loses', title='Derrotas en la UFC'),
    alt.Y('mma_loses', title='Derrotas profesionales en MMA'),
).properties(title='Grafica de Dispersion')
```

Los peleadores tienden a tener iguales o mas derrotas profesionales en MMA que dentro de la UFC.   

```python
alt.Chart(metacritic).mark_bar().encode(alt.X("age", title="Edad"), alt.Y('ufc_wins', title="Victorias en la UFC")).properties(title="Barras con edad y Victorias en la UFC")
```

Esta grafica muestra que el *prime* de los peladores de la UFC se encuentra a los 33 años de edad. Esto se pudiera deber a que tienen uan optima relacion entre experiencia y juventud. A diferencia que tuvieran mas de 40 años con mucha experiencia, pero ya no tanta juventud, y antes de los 20 años lo inverso.

```python
alt.Chart(metacritic).mark_bar().encode(
    alt.X("weight", bin=True, title='Peso en libras'),
    alt.Y('count()', title='Frecuencia')).properties(title="Histograma del peso de Peleadores de la UFC")
```

Las categorias de peso de 120 y 140 libras son las que tienen una mayor cantidad de peleadores que las demas categorias. 

```python
alt.Chart(metacritic).mark_bar().encode(alt.X("gender", title="Sexo"), alt.Y('ufc_no_contests', title="Pelas sin decision en la UFC")).properties(title="Grafica de Barras")
```

Hay una discrepacia entre los peleadores masculinos y femeninos dentro de UFC. Sin embargo no es extraño que sea tan extrema la diferencia debido a la mucho menor a las categorias de peso femenino que el masculino.

```python
alt.Chart(metacritic).mark_bar().encode(
    alt.X("height", bin=True, title='Altura en pulgadas'),
    alt.Y('count()', title='Frecuencia')).properties(title="Histograma de la altura de Peleadores de la UFC")
```

La altura de los peleadores de la UFC sigue el estandar de altura demografica la cual es entre 69 pulgadas o 1.75 metros. Esto significa que no hay un gusto mayor por las MMA en personas bajas o altas.

```python
alt.Chart(metacritic).mark_circle(size=60).encode(
    alt.X('ufc_loses', title='Derrotas en la UFC'),
    alt.Y('ufc_wins', title='Victorias en la UFC'),
).properties(title='Grafica de Dispersion')
```

Dentro de la UFC se concentra mas los peleadores con menos peleas. Esto suena logico puesto que entre mayores peleas tiene un peleador tiende a retirarse del deporte de contacto.

```python
alt.Chart(metacritic).mark_boxplot(extent='min-max').encode(
    alt.X('age:O', title='Edad'),
    alt.Y('ufc_loses:Q', title='Derrotas en la UFC')
).properties(title='Boxplot con minimos y maximos')
```

Se ve una tendencia positiva entre las derrotas de la UFC y la Edad. Y se da un caso extraño a los 35 años, dado que el rango de derrotas es demasiado volatil.  

```python
alt.Chart(metacritic).mark_circle().encode(
    alt.X('mma_wins:O', title='Victorias profesionales en MMA'),
    alt.Y('age:O', title='Edad'),
    alt.Size('sum(mma_loses):Q', title='Sumatoria de Derrotas profesionales en MMA')
).properties(title='Grafico de Burbujas')
```

La edad tambien es un factor en las victorias profesionales de MMA, dando a entender que la edad es un buen factor que determina cuando un peleador se tiene que retirar del deporte o puediera ser un factor de prediccion. Aunque se necesitaria un analisis mas profundo.
